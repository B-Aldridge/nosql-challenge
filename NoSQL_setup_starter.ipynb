5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
# Update latitude and longitude to decimal numbers
update_query_geocode = [
    {"$set": {"geocode.latitude": {"$toDouble": "$geocode.latitude"}}},
    {"$set": {"geocode.longitude": {"$toDouble": "$geocode.longitude"}}}
]

result_geocode = establishments.update_many({}, update_query_geocode)
print("Number of documents updated for latitude and longitude:", result_geocode.modified_count)
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue

# Update RatingValue to integer numbers
update_query_rating = [
    {"$set": {"RatingValue": {"$toInt": "$RatingValue"}}}]
result_rating = establishments.update_many({}, update_query_rating)
print("Number of documents updated for RatingValue:", result_rating.modified_count)
```

```python
# Check that the coordinates and rating value are now numbers
document = establishments.find_one()

# Check data types of latitude, longitude, and RatingValue fields
latitude_type = type(document['geocode']['latitude'])
longitude_type = type(document['geocode']['longitude'])
rating_value_type = type(document['RatingValue'])

print("Data type of latitude:", latitude_type)
print("Data type of longitude:", longitude_type)
print("Data type of RatingValue:", rating_value_type)
```
